```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'notebook', 'grid'])
```

```python
solveA = pd.read_csv('exer3A_out_complete.dat', sep = '\s+').to_numpy().T
solveB = pd.read_csv('exer3B_out_complete.dat', sep='\s+').to_numpy().T
exactA = pd.read_csv('exer3A_out_exact.dat', sep='\s+').to_numpy().T
exactB = pd.read_csv('exer3B_out_exact.dat', sep='\s+').to_numpy().T
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
ax1.set_title(r"Comparativo de $\theta(t)$")
ax2.set_title(r"Comparativo de energia $E(t)$")

ax1.set_xlabel(r"$t \ (s)$")
ax2.set_xlabel(r"$t \ (s)$")
ax1.set_ylabel(r"$\theta \ (rad)$")
ax2.set_ylabel(r"$E \ (N)$")

ax1.plot(exactA[0], exactA[1], label = 'Exato')
ax1.plot(solveA[0], solveA[1], label = 'Euler')
ax1.plot(solveB[0], solveB[1], label = 'Euler-Cromer')

ax1.legend()

#ax2.plot(exactA[0], exactA[2], label = 'Exato')
ax2.plot(solveA[0], solveA[3], label = 'Euler')
ax2.plot(solveB[0], solveB[3], label = 'Euler-Cromer')

ax2.legend()

plt.tight_layout()

plt.savefig('graf3.pdf')
```
